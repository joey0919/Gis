```python
import pandas as pd
```

```python
# csv파일 불러오기
csv = pd.read_csv('서울시 양천구 동물용의료용구판매업 인허가 정보.csv',encoding='cp949')
csv
```

```python
# 데이터프레임 주소값 추출
address= csv['도로명주소']
address
```

```python
# 주소 데이터 깔끔하게 다듬기
new_address = []

for i in range(len(address)):
    if isinstance(address[i], str):
        a = address[i].split(' ')
        new_address.append(" ".join(a[0:4]))
    else:
        new_address.append(address[i])

# 주소 업데이트
address = pd.Series(new_address)
address
```

```python
####### 도로명주소 위도 경도 값으로 바꿔주기 ########
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
#####주소를 위,경도 값으로 변환하기 #####
latitude = []
longitude =[]

for i in address:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
#####Dataframe만들기######
address_df = pd.DataFrame({**csv, '주소':address,'위도':latitude,'경도':longitude})

#df저장
address_df.to_csv('서울시 양천구 동물용의료용구판매업 인허가 정보_가공.csv',encoding='UTF-8-sig', index=False)
```
